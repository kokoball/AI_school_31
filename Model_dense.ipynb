```python
#데이터 부풀리는 작업 

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    zoom_range = 0.2,
    rotation_range=15,
    horizontal_flip = True
)

test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
    TrainImage,
    batch_size = 256,
    shuffle = shuffle,
    target_size = (300,300)
)

test_generator = test_datagen.flow_from_directory(
    TestImage,
    batch_size=50,
    shuffle = shuffle,
    target_size = (300,300)
)
```

```python
trainShape = train_generator.__getitem__(0)[0].shape
testShape = test_generator.__getitem__(0)[0].shape
```

```python
print("Train Shape : ",trainShape)
print("Test Shape : ",testShape)
```

```python
Labels = {'NORMAL':0,'PNEUMONIA':1,'COVID19':2}

def getCode(label):
    return Labels[label]

def getLabel(n):
    for x,c in Labels.items():
        if n==c:
            return x
        
print(getCode('COVID19'))
print(getLabel(1))
```

```python
plt.figure(figsize=(20,10))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(train_generator.__getitem__(0)[0][i])
    plt.title(getLabel(np.argmax(train_generator.__getitem__(0)[1][i])))
```

```python
import glob as gb
import cv2

sizeImage = 300

def getData(Dir,sizeImage):
    X=[]
    y=[]
    for folder in os.listdir(Dir) :
        files = gb.glob(pathname = str(Dir + '/' + folder + '//*.jpg'))
        for file in files :
            picture = cv2.imread(file)
            imageArray = cv2.resize(picture,(sizeImage,sizeImage))
            X.append(list(imageArray))
            y.append(getCode(folder))
    X = np.array(X)
    y = np.array(y)
    
    return X,y
```

```python
X_train, y_train = getData(TrainImage,sizeImage)
X_test,y_test = getData(TestImage,sizeImage)
```

```python
print("X_train Shape : ",X_train.shape)
print("X_test Shape : ",X_test.shape)
```

```python
y_train = to_categorical(y_train,3)
print("y_train",y_train.shape)

y_test = to_categorical(y_test,3)
print("y_test",y_test.shape)
```

## 모델 구축

```python
from tensorflow.keras.applications.densenet import DenseNet201
pre_trained_model = DenseNet201(input_shape = (sizeImage,sizeImage,3), include_top = False,  weights='imagenet')
for layer in pre_trained_model.layers :
    layer.trainable = False
pre_trained_model.summary()
```

```python
from tensorflow.keras import Model


x = tf.keras.layers.Flatten()(pre_trained_model.output)


x = tf.keras.layers.Dense(512, activation='relu')(x)

x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)

x = tf.keras.layers.Dropout(0.4)(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)


x=tf.keras.layers.Dense(3 , activation='sigmoid')(x)
       

model = Model( pre_trained_model.input, x) 

print(model.summary())
model.compile(optimizer='adam', loss="binary_crossentropy",metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau , ModelCheckpoint
lr_reduce = ReduceLROnPlateau(monitor='accuracy', factor=0.1, min_lr=0.0001, patience=1, verbose=1)

filepath="'x_model_dense201.h5'"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
```

```python
epochs = 5
history = model.fit(train_generator,steps_per_epoch=20,callbacks=[lr_reduce,checkpoint] ,epochs=epochs)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'],loc = 'upper left')
plt.show()

plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'],loc = 'upper left')
plt.show()
```

```python
model.evaluate(test_generator)
```

```python
model.save('xray_model.h5')
```
